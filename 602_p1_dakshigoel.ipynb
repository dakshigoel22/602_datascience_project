## Cofounding Variable

**Explanation:**

A confounding variable is an "extra" variable that is associated with both the independent variables and the dependent variable, creating a spurious or misleading relationship between them. It's a hidden factor that can make it seem like X is causing Y, when in reality, the confounder is influencing both.

**Potential Confounder:**

The most significant potential confounding variable in this study is the prompt. The **content and complexity of the prompt** directly influence the linguistic features of the output text (the IVs). For example, a prompt asking for a "poem in the style of Shakespeare" will produce text with very different features than a prompt asking for a "technical summary of a research paper." This influence is independent of which LLM is used. If one LLM in the dataset was predominantly used for creative prompts while another was used for technical ones, the model might incorrectly learn that "poetic language" is a feature of the first LLM, when it's actually a feature of the prompt's domain.

**How to Address It:**

Stratified Analysis: Ensure that the dataset was created by giving the exact same set of prompts to every LLM and the human writers. This design neutralizes the prompt as a confounder.

Feature Engineering: If prompts are varied, they can be categorized (e.g., 'Creative', 'Technical', 'News'). This prompt category can then be included as a control variable in the model to account for its effect.

Acknowledge as a Limitation: If the prompts are too diverse to categorize and were not applied systematically, this must be acknowledged as a limitation of the study, as it could influence the results

## **6.1 WHY ?**

I chose the dataset for the following reasons:

1. **Quality :** The dataset's primary strength is its use of a broad array of powerful and relevant LLMs, including ChatGLM, Dolly, ChatGPT-turbo, GPT4All, StableLM, and Claude. In a rapidly evolving field, using a dataset with state-of-the-art models ensures the research findings are timely and applicable to the current technological landscape. Other dataset includes the TuringTest Benchmark dataset, but it was last updated in 2021, contains outdated LLMs and does not consist of information about prompts.

2. **Consistency :** The dataset provides both human-written and AI-generated text for the same set of prompts. This parallel structure is a critical feature for a methodologically sound comparison. By having every author respond to the same prompt, the dataset design isolates the author's unique style as the key variable under investigation. This minimizes the influence of the prompt's topic, which would otherwise act as a significant confounding variable, ensuring a fair comparison of each author's stylistic fingerprint.

3. **Variety :** The prompts are consistently sourced from three distinct domains: essays, Wikipedia articles, and Reuters news reports. This enables a deeper analysis of how stylistic markers might vary across different contexts, directly supporting the investigation of hypotheses related to domain influence.


However, at the same time, it is also important to address the limitations of the MGT BENCH dataset, particularly the lack of information on prompts and generation parameters. For this study, I am assuming that the parameters are kept similar across models and across prompts.

```python
# Columns that contain labels
label_columns = ["human", "ChatGPT-turbo", "Claude", "ChatGLM", "Dolly", "ChatGPT", "GPT4All", "StableLM"]

import pandas as pd

# Convert the llm_data into long format using melt
data = llm_data.melt(
    id_vars=["id","author" ,"prompt"],
    value_vars=label_columns,  # e.g., ["human","ChatGPT","Claude",...]
    value_name="text",
    var_name="label"
)

# Text length features :
data["char_length"] = data["text"].str.len()
data["word_count"] = data["text"].str.split().str.len()
data["sentence_count"] = data["text"].str.split('.').str.len()

```
